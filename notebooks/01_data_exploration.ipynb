```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = {
    "engine_rpm": [800, 1500, 3000, 2000, 3500, 1200],
    "vehicle_speed": [0, 40, 100, 60, 120, 30],
    "engine_temp": [90, 95, 110, 100, 115, 92],
    "intake_pressure": [30, 45, 60, 50, 65, 40],
    "fault_type": [
        "No Fault",
        "Oxygen Sensor Fault",
        "Engine Misfire",
        "Emission System Fault",
        "Engine Misfire",
        "Oxygen Sensor Fault"
    ]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
df.describe()
```

```python
sns.countplot(data=df, x="fault_type")
plt.xticks(rotation=45)
plt.title("Distribution of Automotive Fault Types")
plt.show()
```

```python
df.to_csv("../data/processed/cleaned_sensor_data.csv", index=False)
```
